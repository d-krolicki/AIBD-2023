```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import numpy as np
```

## Przygotowanie danych

```python
NAME = "Dariusz"
SURNAME = "Królicki"
FILE_NO = len(NAME) + len(SURNAME) % 16
FILENAME = ""
FOLDER_PATH = os.path.join("..", "Original Data")
for file in os.listdir(FOLDER_PATH):
    if file.startswith(str(FILE_NO)):
        FILENAME = file
FILE_PATH = os.path.join(FOLDER_PATH, FILENAME)
```

```python
df = pd.read_csv(FILE_PATH)
```

```python
keys = [key for key in df]
keys = keys[1:]
print(keys)
```

## Dane o ilości dni, które upłynęły miedzy datą zakupu a wystawieniem oceny

```python
bins_for_days_since_bought = max(df[keys[0]])
```

```python
fig0 = df[keys[0]].hist(bins=bins_for_days_since_bought, figsize=(4,4), xlabelsize=10, ylabelsize=10)
fig0.set_xlabel('Dni od zakupu', fontsize=10)
fig0.set_ylabel('Ilość wyników', fontsize=10)
plt.show()
```

## Dane dot. marki zakupionego sprzętu

```python
counts = df[keys[1]].value_counts()
fig1 = counts.plot.bar()
fig1.set_xlabel("Nazwa marki", fontsize=10)
fig1.set_ylabel("Ilość zakupionych produktów", fontsize=10)
plt.show()
```

## Dane dotyczące wieku kupującego

```python
bins_for_age = int(max(df[keys[2]]) // 5)
```

```python
fig2 = df[keys[2]].hist(bins=bins_for_age)
fig2.set_xlabel("Wiek kupującego", fontsize=10)
fig2.set_ylabel("Ilość rekordów", fontsize=10)
plt.show()
```

## Płeć kupującego

```python
counts = df[keys[3]].value_counts()
fig1 = counts.plot.bar()
fig1.set_xlabel("Płeć kupującego", fontsize=10)
fig1.set_ylabel("Ilość rekordów", fontsize=10)
plt.show()
```

## Ocena

```python
bins_for_rating = len(df.value_counts(df[keys[4]]))
ticks = np.linspace(0, 5, bins_for_rating)
```

```python
fig2 = df[keys[4]].hist(bins=bins_for_rating)
plt.xticks(ticks)
fig2.set_xlabel("Ocena", fontsize=10)
fig2.set_ylabel("Ilość rekordów", fontsize=10)
plt.show()
```
