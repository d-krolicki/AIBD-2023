```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def f(x:np.float64):
    return x**2 + 5
```

```python
x1 = np.linspace(-1, 1, 100)
x2 = np.linspace(-6, 6, 100)
x3 = np.linspace(0, 5, 100)

y1 = np.array([f(x) for x in x1])
y2 = np.array([f(x) for x in x2])
y3 = np.array([f(x) for x in x3])
```

```python
fig, axs = plt.subplots(1, 3, figsize=(12,5))
axs[0].plot(x1, y1, color='red')
axs[0].set_title("Graph from -1 to 1")
axs[1].plot(x2, y2, color='green')
axs[1].set_title("Graph from -6 to 6")
axs[2].plot(x3, y3, color='blue')
axs[2].set_title("Graph from 0 to 5")

for ax in axs:
    ax.set_xlabel("Arguments")
    ax.set_ylabel("Values")
    ax.legend(['Value'], loc="lower right")
```
